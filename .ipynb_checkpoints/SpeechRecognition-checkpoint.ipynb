```python
def simpson(audio_data, start, step):
    h = 1 / rate
    a = start
    b = start + 2 * step
    
    if (a == b): return 0
    
    return (h * (audio_data[a][1] + 4 * audio_data[a+1][1] + audio_data[a+2][1]) / 3) + simpson(audio_data, a+2, step - 1) 
    
```

```python
def trapezoid(audio_data, start, end):
    h = 1 / rate
    a = start
    b = end
    
    if (a == b): return 0
    
    return (h / 2) * (audio_data[a][1] + audio_data[a+1][1]) + trapezoid(audio_data, a+1, end)
```

```python
def leftpoint_rule(audio_data, start, end):
    h = 1 / rate
    a = start
    b = end
    if (a == b): return 0
    return h * audio_data[a][1] + leftpoint_rule(audio_data, a+1, b)
```

```python
print(leftpoint_rule(audio_data, 0, 20000))
```

```python
import math
def window(x, n):
    return 0.54 - 0.46 * math.cos(2 * math.pi * x / (n - 1))

def get_freq_with_simpson(audio_data, s):
    h = 1 / rate
    a = 0
    b = 40000

    sum = 0
        
    while (a != b):
    
        y0 = window(a, b) * math.e**(-s*a*h) * audio_data[a][1]
        y1 = window(a+1, b) * math.e**(-s*(a+1)*h) * audio_data[a+1][1]
        y2 = window(a+2, b) * math.e**(-s*(a+2)*h) * audio_data[a+2][1]
        
        # print("0", str(y0), "1", y1, "2", y2)
    
        sum += h * (y0 + 4 * y1 + y2) / 3
        a += 2
    
    return sum
    
def get_freq_data(audio_data, step):
    freq_data = []
    s = 20
    while (s <= 20000):
        freq_data.append(get_freq_with_simpson(audio_data, s))
        s += step
    return freq_data
```

```python
freq_data = get_freq_data(audio_data, 10)
```

```python
def get_model_data(freq_data, interval):
    model_data = [0]*(int(math.ceil(len(freq_data) / interval)))
    # print(len(model_data))
    pointer = 0
    i = 0
    while pointer < len(freq_data):
        model_data[i] += freq_data[pointer]
        pointer += 1
        if pointer % interval == 0:
            i += 1
    return model_data
```

```python
model_data2 = get_model_data(freq_data, 100)
```
